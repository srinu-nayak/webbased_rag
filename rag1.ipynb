```python
import os
from langchain_community.document_loaders import WebBaseLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from dotenv import load_dotenv

load_dotenv()
os.environ["OPENAI_API_KEY"] = os.environ["OPENAI_API_KEY"]
```

```python
#loading data from webpage
page_url = "https://www.educosys.com/course/genai"
loader = WebBaseLoader(web_paths=[page_url])
docs = loader.load()
print(docs[0].page_content)
```

```python
#text-splitting
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=100,
    chunk_overlap=20,
)
splits = text_splitter.split_documents(docs)
```

```python
#Embeddings
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import  Chroma

vector_db = Chroma.from_documents(splits, OpenAIEmbeddings())
retriver = vector_db.as_retriever()

print(vector_db._collection.get())
```

```python
print("\n Collection1 - ",  vector_db._collection.get(ids="91854e6f-2753-48a2-92f3-448599f64ec5", include=["embeddings", "documents"]))

print("\n Collection2 - ",  vector_db._collection.get(ids="7e0f5a6f-d97a-4f8c-a4c9-d38b06966f23", include=["embeddings", "documents"]))
```

```python
#rag query
from langchain import hub
from langchain_core.runnables import RunnableLambda, RunnablePassthrough
from langchain_openai import OpenAI
from langchain_core.output_parsers import StrOutputParser

llm = OpenAI()
prompt = hub.pull("rlm/rag-prompt")
```

```python
def format_docs(docs):
    return "\n".join(doc.page_content for doc in docs)
```

```python
def print_prompt(prompt):
    print("Prompt - ")
    return prompt
```

```python
rag_chain = (
    {"context": retriver | format_docs, "question": RunnablePassthrough()}
    | RunnableLambda(print_prompt)
    | prompt
    | llm
    | StrOutputParser()
)

rag_chain.invoke("what is genai?")


```

```python
from ragas import evaluate
from ragas.metrics import faithfulness, answer_relevancy, context_recall, context_precision
from datasets import Dataset

# Example evaluation questions
eval_questions = [
    "What is GenAI?",
    "Who can benefit from this course?",
    "What topics are covered in the course?",
]

# Collect predictions
samples = []
for q in eval_questions:
    retrieved_docs = retriver.get_relevant_documents(q)
    response = rag_chain.invoke(q)

    samples.append({
        "question": q,
        "answer": response,
        "contexts": [d.page_content for d in retrieved_docs]
    })

# Convert into HF Dataset
dataset = Dataset.from_list(samples)
```

```python
result = evaluate(
    dataset=dataset,
    metrics=[faithfulness, answer_relevancy, context_recall, context_precision]
)

print(result)
```
